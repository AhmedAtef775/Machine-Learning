```python
#delete rows (outliers)
del_outliers = df[(df['total_rooms'] > 15000) &  (df['population'] > 12000)]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = del_outliers.drop(['ocean_proximity' , 'gender' , 'median_house_value'] , axis= 1).values
```

```python
y = del_outliers['median_house_value'].values
```

```python
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size = 0.25 , random_state = 40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
#evaluation
from sklearn.metrics import r2_score
```

```python
acc = r2_score(y_test , y_pred) * 100
acc
```
